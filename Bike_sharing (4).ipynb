```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
# Number of bootstrap iterations
num_iterations = 100
# Lists to store MSE and coefficients from each iteration
mse_values = []
coefficients_list = []
for i in range(num_iterations):
    # Create a random training set using bootstrap resampling
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=None, shuffle=True)
    alpha = 0.01
    lasso = Lasso(alpha=alpha)
    lasso.fit(x_train, y_train)
    y_pred = lasso.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    coefficients_list.append(lasso.coef_)
mean_mse = np.mean(mse_values)
std_mse = np.std(mse_values)
print(f"Mean of MSE across {num_iterations} iterations: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
average_coefficients = np.mean(coefficients_list, axis=0)
print("Average Lasso Coefficients:", average_coefficients)
```
